```python
import re
def remove_emoji(text: str):
    regex_pattern = re.compile(pattern="["
u"\U0001F600-\U0001F64F" # emoticons
u"\U0001F300-\U0001F5FF" # symbols & pictographs
u"\U0001F680-\U0001F6FF" # transport & map symbols
u"\U0001F1E0-\U0001F1FF" # flags (iOS)
u"\U00002500-\U00002BEF" # chinese char
u"\U00002702-\U000027B0"
u"\U00002702-\U000027B0"
u"\U000024C2-\U0001F251"
u"\U0001f926-\U0001f937"
u"\U00010000-\U0010ffff"
u"\u2640-\u2642"
u"\u2600-\u2B55"
u"\u200d"
u"\u23cf"
u"\u23e9"
u"\u231a"
u"\ufe0f" # dingbats
u"\u3030"
"]+", flags=re.UNICODE)
    
    return regex_pattern.sub(r'', text)
```

### Semantic analysis - positive/negative reviews, bigrams, trigrams

```python
# drop data with missing values in the 'Review' column
app_reviews = app_reviews.dropna(axis=0, subset=['content'])
```

```python
# convert the relevant column to lowercase since nltk works with lists, not dataframes
app_reviews['content'] = app_reviews.content.str.lower()

# create a list of the reviews from the 'Reviews' column
reviews = app_reviews['content'].tolist()
```

```python
reviews[0:2]
```

```python
# tokenise the words
from nltk.tokenize import word_tokenize

content_tokens = []
for word in reviews:
    content_tokens.append(word_tokenize(word))
```

```python
# content_tokens[0]
```

```python
# app_reviews.head()
```

```python
import nltk
nltk.download('stopwords')

from nltk.corpus import stopwords

# use English stopwords
stops = set(stopwords.words("english"))
```

```python
# Using the stopwords above, create a new list with the stopwords removed from word_tokens
filtered_content = []
for word in content_tokens:
    filtered_content.append([w for w in word if not w in stops])
    
# adding all these new tokens to a new column in df
app_reviews['filtered_content'] = filtered_content
```

```python
app_reviews.head()
```

```python
# Finally, put all the tokens back together as a text:
import string
joined_tokens = []
for word in filtered_content:
    z = ",".join(word) # join all the stemmed words back together on the commas
    z = z.replace(",", " ") # replace all the commas with spaces
    # remove the punctuation from the text:
    joined_tokens.append(z.translate(str.maketrans('', '', string.punctuation)))
    
# add this new and reformed text to the data frame
app_reviews['Joined token reviews'] = joined_tokens
```

```python
# app_reviews.head()
```

```python
from sklearn.feature_extraction.text import CountVectorizer, TfidfVectorizer

# set vectorizer
vectorizer = CountVectorizer()

# array of word counts
document_vec = vectorizer.fit_transform(joined_tokens)

# convert to dataframe
df2 = pd.DataFrame(document_vec.toarray(), columns=vectorizer.get_feature_names())

# set a threshold to drop infrequent words
threshold = 0.1

# drop words based on the threshold
df2 = df2.drop(df2.mean()[df2.mean() < threshold].index.values, axis=1) # axis=1 for column mean
```

```python
# join the two datasets or dataframes together
app_reviews = app_reviews.join(df2, how='left')
```

```python
app_reviews.head()
```

```python
# We use df1 as it contains no links
# Find all the bigrams

from nltk import bigrams

# Find the best 'best' bigrams / trigrams

from nltk.collocations import *
bigram_measures = nltk.collocations.BigramAssocMeasures()
trigram_measures = nltk.collocations.TrigramAssocMeasures()
quadgram_measures = nltk.collocations.QuadgramAssocMeasures()

# create a single string that combines 200 reviews
# words is the list of reviews created earlier
long_string = word_tokenize(",".join(joined_tokens[0:250]))

# Bigrams
finder = BigramCollocationFinder.from_words(long_string)
# only bigrams that appear >= 2x
finder.apply_freq_filter(2)
# return the 10 bigrams with the highest likelihood
print(finder.nbest(bigram_measures.likelihood_ratio, 10))
```

```python
# Trigrams
finder = TrigramCollocationFinder.from_words(long_string)
# only bigrams that appear >= 2x
finder.apply_freq_filter(2)
# return the 10 trigrams with the highest likelihood
print(finder.nbest(trigram_measures.likelihood_ratio, 10))
```

```python
# Quadgrams
quadgram_measures = nltk.collocations.QuadgramAssocMeasures()

finder = QuadgramCollocationFinder.from_words(long_string)
# only bigrams that appear >= 2x
finder.apply_freq_filter(2)
# return the 10 trigrams with the highest likelihood
print(finder.nbest(quadgram_measures.likelihood_ratio, 10))
```

## Positive/Negative reviews

```python
from textblob import TextBlob

# calculate polarity and subjectivity
app_reviews['Polarity'] = app_reviews['content'].apply(lambda review: TextBlob(str(review)).sentiment.polarity)
app_reviews['Subjectivity'] = app_reviews['content'].apply(lambda review: TextBlob(str(review)).sentiment.subjectivity)

# sort by polarity (low to high)
sorted_df = app_reviews.sort_values(by=['Polarity'])

# print top 5 positive and negative
print("Most positive #8 tweets ")
print(sorted_df['content'].tail(8))
print("\n") # print line break
print("Most negative #8 tweets ")
print(sorted_df['content'].head(8))
```
